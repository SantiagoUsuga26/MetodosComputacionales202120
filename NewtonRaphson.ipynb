# Legendre

```python
def GetLegendre(n):
    
    x = sym.Symbol('x',Real=True)
    y = sym.Symbol('y',Real=True)
    
    y = (x**2-1)**n
    
    p = sym.diff(y,x,n)/(2**n * np.math.factorial(n))
    
    return p
```

```python
GetLegendre(2)
```

```python
Legendre = []

n=7
for i in range(n+1):
    Legendre.append(GetLegendre(i))
```

```python
#Legendre
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(1,1,1)

t = np.linspace(-1,1,100)
x = sym.Symbol('x',Real=True)

for i in range(1,n+1):
    pn = sym.lambdify([x],Legendre[i],'numpy')
    ax.plot(t,pn(t),label=r'$Order=%.0f$'%(i))
    
ax.axhline(y=0,ls='--',color='r')

ax.legend()
```

```python
def GetLaguerre(n):
    
    x = sym.Symbol('x',Real=True)
    y = sym.Symbol('y',Real=True)
    
    y = sym.exp(-x)*x**n
    
    p = sym.exp(x)*sym.diff(y,x,n)/(np.math.factorial(n))
    
    return p
```

```python
GetLaguerre(1)
```

```python
Laguerre = []

n=7
for i in range(n+1):
    Laguerre.append(GetLaguerre(i))
```

```python
Laguerre
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(1,1,1)

t = np.linspace(0,10,100)
x = sym.Symbol('x',Real=True)

for i in range(1,n+1):
    pn = sym.lambdify([x],Laguerre[i],'numpy')
    ax.plot(t,pn(t),label=r'$Order=%.0f$'%(i))
    
ax.axhline(y=0,ls='--',color='r')

ax.legend()
```
