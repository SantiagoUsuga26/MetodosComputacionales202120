```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,20)
h  = x[1]-x[0]
print(x[1]-x[0])
```

```python
sin = lambda x: np.sin(x)
```

```python
f = sin(x)
```

```python
def DerivativeR(x,f,h):
    return (f(x+h)-f(x))/h
```

```python
def DerivativeL(x,f,h):
    return (f(x)-f(x-h))/h
```

```python
def DerivativeC(x,f,h):
    return (f(x+h)-f(x-h))/(2*h)
```

```python
DerivadaR = DerivativeR(x,sin,h)
DerivadaL = DerivativeL(x,sin,h)
DerivadaC = DerivativeC(x,sin,h)
```

```python
fig = plt.figure(figsize=(7,3))
ax = fig.add_subplot(2,2,1)
ax1 = fig.add_subplot(2,2,2)
ax2 = fig.add_subplot(2,2,3)


ax.scatter(x,f)
ax1.scatter(x,DerivadaR)
ax1.scatter(x,DerivadaL)
ax1.scatter(x,DerivadaC)

ax1.scatter(x,np.cos(x))
ax2.scatter(x,np.abs(np.cos(x)-DerivadaR))
ax2.scatter(x,np.abs(np.cos(x)-DerivadaL))
ax2.scatter(x,np.abs(np.cos(x)-DerivadaC))
```
