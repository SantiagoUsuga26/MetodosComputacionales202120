```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
from mpl_toolkits.mplot3d import axes3d
```

$$  I = \int_{0}^{2\pi} \int_{0}^{2} \int_{0}^{1} \rho z dV = \frac{4}{3} \pi$$ 

```python
def CreateCylinder( N, zmin, zmax, R ):
    
    X = np.zeros(N)
    Y = np.zeros_like(X)
    Z = np.zeros_like(X)
    
    
    for i in tqdm(range(N)):
        
        u = np.random.rand()
        r = R*u**(1/2)
        
        phi = np.random.uniform(0,2*np.pi)
        z = np.random.uniform(zmin,zmax)
        
        X[i] = r*np.cos(phi)
        Y[i] = r*np.sin(phi)
        Z[i] = z
    
    
    return X,Y,Z
```

```python
X,Y,Z = CreateCylinder( 10000, 0,2,1 )
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1, projection = '3d')

#configurar límites de los ejes
ax.set_xlim3d(-1, 1)
ax.set_ylim3d(-1, 1)
ax.set_zlim3d(0, 2)

ax.view_init(10, 60)

ax.scatter(X,Y,Z,color='b')
```

```python
def f(r,z):
    return r*z
```

```python
def GetPoints(zmin,zmax,R):
    
    u = np.random.rand()
    r = R*u**(1/2)    
    z = np.random.uniform(zmin,zmax)
    
    return r,z
```

```python
N = int(1e6)
Sample = np.zeros(N)

for i in range(N):
    r,z = GetPoints(0,2,1)
    Sample[i] = f(r,z)
    
```

```python
Integral = 2*np.pi * np.average( Sample )
```
